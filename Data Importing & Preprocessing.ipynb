## Treating Variables in Python

```python
df.info()
```

```python
df['n_hos_beds'] = df['n_hos_beds'].fillna(df['n_hos_beds'].mean())
```

```python
np.percentile(df.n_hot_rooms, [99])
```

```python
np.percentile(df.n_hot_rooms, [99])[0] #fetching the first element of 99 percentile array
```

```python
uv = np.percentile(df.n_hot_rooms, [99])[0]
```

```python
df[(df.n_hot_rooms > uv)] #where the column is greater than the 99 percentile values
```

```python
df.n_hot_rooms[(df.n_hot_rooms>3*uv)] = 3*uv #treating the outliers 
```

```python
np.percentile(df.rainfall, [1])[0]
```

```python
lv = np.percentile(df.rainfall, [1])[0]
```

```python
df[(df.rainfall < lv)] #identifying values lower than 1 percentile
```

```python
df.rainfall[(df.rainfall<0.3*lv)]=0.3*lv
```

## Applying a Simple Linear Regression Model

```python
import statsmodels.api as sn
```

```python
X = sn.add_constant(df['room_num']) #create variable X & add constant
```

```python
lm = sn.OLS(df['price'], X).fit() #OLS = Ordinary Least Square, price is our dependent variable, fitting our model
```

```python
lm.summary() #getting the summary of the model
```

```python
from sklearn.linear_model import LinearRegression #most commonly used Linear regression method in python
```

```python
import numpy as np
```

```python
y = df['price'] #dependent variable
```

```python
X = df[['room_num']] #2d array
```

```python
LR = LinearRegression()
```

```python
LR.fit(X, y)
```

```python
print(LR.intercept_, LR.coef_) #printing the attributes of our linear regression model 
```

### Test Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X1, X2, y1, y2 = train_test_split(X_all, y_all, test_size = .2, random_state = 0)
```

```python
print(X1.shape, X2.shape, y1.shape, y2.shape)
```

```python
LR_A = LinearRegression()
```

```python
LR_A.fit(X1, y1)
```

```python
y2_a = LR_A.predict(X2)
```

```python
y1_a = LR_A.predict(X1)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y2, y2_a)
```

```python
r2_score(y1, y1_a)
```
